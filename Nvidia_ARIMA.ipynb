```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('NVDA Research.csv',index_col='Date',parse_dates=True)
```

```python
df=df.dropna()
print('Shape of data',df.shape)
df.head()
```

```python
df['Adj Close'].plot(figsize=(12,5))
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def ad_test(dataset):
  dftest=adfuller(dataset,autolag="AIC")
  print("1. ADF : ",dftest[0])
  print("2. P-value : ",dftest[1])
  print("3. Num of Lags : ", dftest[2])
  print("4. Num of observations used for ADF Regression and critical Values Calculation : ", dftest[3])
  print("5. critical values : ")
  for key, val in dftest[4].items():
    print("\t",key, ":", val)
```

```python
ad_test(df['Adj Close'])
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
print(df.shape)
train=df.iloc[:-145]
test=df.iloc[-145:]
print(train.shape,test.shape)
```

```python
model= ARIMA(train['Adj Close'],order=(5,2,0))
model=model.fit()

print(model.summary())
```

```python
start = len(train)
end = len(train)+len(test)-1
pred=model.predict(start=start,end=end,type='levels')
print(pred)
```

```python
pred.plot(legend=True)
test['Adj Close'].plot(legend=True)
```

```python
test['Adj Close'].mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(pred,test['Adj Close']))
print(rmse)
```
